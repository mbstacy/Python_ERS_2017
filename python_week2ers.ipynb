### Create Function to plot data

1. Create plot and export out to file

```python
def myplot(x,y,style,title,xlabel,ylabel,figsize=(20,5),filename="test.png"):
    fig, axes = plt.subplots()
    ax=y.plot(ax=axes,figsize=figsize,title=title,style=style)
    plt.xticks(rotation=90)
    plt.xlabel(xlabel)
    plt.ylabel(ylabel)
    # set xaxis major Dateformater
    plt.gca().xaxis.set_major_formatter(mdates.DateFormatter('%Y-%m-%d'))
    # set grid lines
    ax = plt.gca().grid(True)
    fig.savefig(filename,dpi=300, bbox_inches='tight')
```

```python
x=rf_data.index
y= rf_data.ix[:,'GF_NDVI':'GF_EVI']
myplot(x,y ,'',"Vegitation Index","Date","Vegitation Index",
       figsize=(20,10),
      filename="test.jpg")
```

## Pandas apply

```python
def vi_classification(exp):
    if exp < 0.25:
        # This observation is low 
        return "Low VI"
    elif exp >= 0.25 and exp<= 0.75:
        # This observation is Mid Range
        return "Mid Range VI"
    elif exp >= 0.75:
        # This observation is High Range
        return "High Range VI"
    else:
        # This observation has bad data
        return "NA"
    #return "NA"


```

```python
rf_subset.columns
```

```python
rf_subset['VI_Classification']=rf_subset['GF_NDVI'].apply(vi_classification)
```

```python
rf_subset
```

### Row level apply function


Send entire row to the function.

```python
def vi_classification_row(dr):
    if dr['GF_NDVI'] >0.5 and dr['GF_EVI']<0.5:
        return "Set1"
    else:
        try:
            result = dr['GF_NDVI'] * dr['GF_EVI']
        except:
            result = None
        return result
```

```python
rf_subset['set']=rf_subset.apply(vi_classification_row,axis=1)
```

```python
rf_subset
```

```python
myplot(rf_data.index, rf_data['GF_NDVI'],'g-',"NDVI","Date","NDVI")
```

### Export to Script 

We want to export our myplot function to a script.

1. Copy function and add to .py file
2. if __name__ == '__main__': at bottom of script
3. sys.argv to collect command line arguments
4. $ python filename args 

Demonstration of script file!!!!

See below for entire function parts.

```python
import matplotlib.pyplot as plt
import matplotlib.dates as mdates
import pandas as pd
import sys

def myplot(x,y,style,title,xlabel,ylabel,figsize=(20,5),filename="test.png"):
    fig, axes = plt.subplots()
    ax=y.plot(ax=axes,figsize=figsize,title=title,style=style)
    plt.xticks(rotation=90)
    plt.xlabel(xlabel)
    plt.ylabel(ylabel)
    # set xaxis major Dateformater
    plt.gca().xaxis.set_major_formatter(mdates.DateFormatter('%Y-%m-%d'))
    # set grid lines
    ax = plt.gca().grid(True)
    fig.savefig(filename,dpi=300, bbox_inches='tight')
```
